# Financial Impact Estimation: Quantity Sold vs Total Price

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
Finacial_impact_data = df_cleaned.groupby('product_id').agg({
    'qty' : 'sum',
    'total_price' : 'sum'
}).reset_index()
Finacial_impact_data.head()
```

```python
total_qty_sold =Finacial_impact_data ['qty'].sum()
total_revenue = Finacial_impact_data['total_price'].sum()
total_qty_sold, total_revenue 
```

```python
#top performance product 
top_products_by_revenue = Finacial_impact_data.nlargest(3,'total_price')

top_products_by_revenue
```

we can see product health2 generated the highest revenue of $63895 with a total units sold 195
this will show the each products peroformance 

```python
#low performance product 
low_product_by_revenue =Finacial_impact_data.nsmallest(3,'total_price')
low_product_by_revenue
```

```python
#Revenue per Unit:
Finacial_impact_data['revenue_per_unit']=Finacial_impact_data['total_price'] / Finacial_impact_data['qty']
# Get the top 5 products by revenue per unit
top_5_revenue_per_unit = Finacial_impact_data.nlargest(5, 'revenue_per_unit')
top_5_revenue_per_unit 

```

```python
# Get the bottom 5 products by revenue per unit
low_5_revenue_per_unit = Finacial_impact_data.nsmallest(5, 'revenue_per_unit')
low_5_revenue_per_unit
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the width of the bars
bar_width = 0.35

# Define the positions for the bars on the x-axis
product_indices = np.arange(len(Finacial_impact_data['product_id']))

# Scale the quantity sold by multiplying it by 500
scaled_qty = Finacial_impact_data['qty'] * 500  # Adjust the scale factor as needed

# Create the figure and axis
plt.figure(figsize=(12, 6))

# Create bars for Total Price
plt.bar(product_indices, Finacial_impact_data['total_price'], width=bar_width, label='Total Price', color='blue', alpha=0.6)

# Create bars for Scaled Quantity Sold, offset by the width of the bars
plt.bar(product_indices + bar_width, scaled_qty, width=bar_width, label='Quantity Sold', color='orange', alpha=0.6)

# Add labels and title
plt.xlabel('Product ID')
plt.ylabel('Total Price / Scaled Quantity Sold')
plt.title('Financial Impact: Quantity Sold vs Total Price (Side-by-Side)')

# Add x-ticks with the product IDs, adjusting for both sets of bars
plt.xticks(product_indices + bar_width / 2, Finacial_impact_data['product_id'], rotation=90, fontsize=8)

# Add a legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

# Price Competitiveness: Compare the company’s prices against competitors' (comp_1, comp_2, comp_3).

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# define the width of the bars 
bar_width = 0.15

# define the position for the bars on the x_axis 
product_indices = np.arange(len( df_cleaned['product_id']))

#create the figure and axis 
plt.figure(figsize =(16,8))

#plot the compony's unit price
plt.bar(product_indices, df_cleaned['unit_price'],width=bar_width,label ='company price', color ='blue',alpha=0.6)
# Plot competitors' prices (comp_1, comp_2, comp_3), shifting their positions for side-by-side comparison
plt.bar(product_indices + bar_width, df_cleaned['comp_1'],width=bar_width,label ='competitor 1', color ='orange',alpha=0.6)
plt.bar(product_indices + 2*bar_width, df_cleaned['comp_2'],width=bar_width,label ='competitor 2', color ='Green',alpha=0.6)
plt.bar(product_indices +3*bar_width, df_cleaned['comp_3'],width=bar_width,label ='competitor 3', color ='red',alpha=0.6)

#add x-ticks with the product IDS, adjudting for all the bar groups
plt.xticks(product_indices[::5] + 1.5*bar_width, df_cleaned['product_id'][::5], rotation =90, fontsize =8)

#add a legend 
plt.legend()

#Display the plot
plt.tight_layout()
plt.show()
```

```python
# Select relevant columns for price comparison
lineplot_data = df_cleaned[['product_id', 'unit_price', 'comp_1', 'comp_2', 'comp_3']].dropna()

# Create a line plot
plt.figure(figsize=(12, 6))

# Plot the company prices
plt.plot(lineplot_data['product_id'], lineplot_data['unit_price'], label='Company Price', marker='o')

# Plot competitors' prices
plt.plot(lineplot_data['product_id'], lineplot_data['comp_1'], label='Competitor 1', marker='o')
plt.plot(lineplot_data['product_id'], lineplot_data['comp_2'], label='Competitor 2', marker='o')
plt.plot(lineplot_data['product_id'], lineplot_data['comp_3'], label='Competitor 3', marker='o')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Add labels and title
plt.xlabel('Product ID')
plt.ylabel('Price')
plt.title('Line Plot: Price Comparison for Company vs Competitors')

# Add a legend
plt.legend()

# Display the line plot
plt.tight_layout()
plt.show()


```

```python
#Calculate the average price for the company and compititors
# Calculate the average price for the company and competitors
avg_company_price = df_cleaned['unit_price'].mean()
avg_comp_1_price = df_cleaned['comp_1'].mean()
avg_comp_2_price = df_cleaned['comp_2'].mean()
avg_comp_3_price = df_cleaned['comp_3'].mean()

# Display the average prices
avg_company_price, avg_comp_1_price, avg_comp_2_price, avg_comp_3_price
```

```python
# Correctly calculate price deviations for each competitor
df_cleaned['competitor_1'] = df_cleaned['unit_price'] - df_cleaned['comp_1']
df_cleaned['competitor_2'] = df_cleaned['unit_price'] - df_cleaned['comp_2']
df_cleaned['competitor_3'] = df_cleaned['unit_price'] - df_cleaned['comp_3']  # Correct competitor_3 column

# Get the top 5 products where our company price is the highest relative to Competitor 1
top_5_high_products= df_cleaned.nlargest(5,'competitor_1')

top_5_high_products[['product_id', 'unit_price', 'comp_1', 'competitor_1','comp_2', 'competitor_2','comp_3', 'competitor_3']]



```

```python
# Get the least  5 products where our company price is the cheaper relative to Competitor 
top_5_cheap_products = df_cleaned.nsmallest(5, 'competitor_1')

top_5_cheap_products[['product_id', 'unit_price', 'comp_1', 'competitor_1','comp_2', 'competitor_2','comp_3', 'competitor_3']]
```

```python
# Assuming 'month_year' is available for each product, we can group by that
price_trends = df_cleaned.groupby('month_year').agg({
    'unit_price': 'mean',
    'comp_1': 'mean',
    'comp_2': 'mean',
    'comp_3': 'mean'
}).reset_index()

# Plot the price trends over time
plt.figure(figsize=(12, 6))
plt.plot(price_trends['month_year'], price_trends['unit_price'], label='Company Price', marker='o')
plt.plot(price_trends['month_year'], price_trends['comp_1'], label='Competitor 1 Price', marker='o')
plt.plot(price_trends['month_year'], price_trends['comp_2'], label='Competitor 2 Price', marker='o')
plt.plot(price_trends['month_year'], price_trends['comp_3'], label='Competitor 3 Price', marker='o')

# Add labels and title
plt.xlabel('Month-Year')
plt.ylabel('Average Price')
plt.title('Price Trends Over Time')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
