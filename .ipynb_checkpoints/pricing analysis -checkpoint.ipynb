```python
# Define the path as a raw string
pricing_data_load = Path(r"pricing analysis/pricing analyis/retail pricing.csv")


```

```python
# Read the data and store it into a pandas DataFrame
price_data = pd.read_csv(pricing_data_load)
```

```python
price_data.head()
```

```python
#Check basic info 
price_data.info()
```

```python
price_data.describe()
```

```python
# Check for missing values
missing_values = price_data.isnull().sum()
missing_values
```

```python
# Select relevant columns based on pricing analysis needs
relevant_columns = ['product_id', 'product_category_name', 'month_year', 'qty', 'total_price', 'unit_price', 
                    'volume','month','year','comp_1', 'comp_2', 'comp_3', 'lag_price']
relevant_columns
```

```python
# Filter the DataFrame to keep only the relevant columns
df_cleaned = price_data[relevant_columns]
df_cleaned.head()
```

```python
# Check the summary of the cleaned dataset
df_cleaned.info(), missing_values
```

```python
import matplotlib.pyplot as plt
```

```python
# Create additional metrics: price difference (vs competitors)
df_cleaned['price_diff_comp1'] = df_cleaned['unit_price'] - df_cleaned['comp_1']
df_cleaned['price_diff_comp2'] = df_cleaned['unit_price'] - df_cleaned['comp_2']
df_cleaned['price_diff_comp3'] = df_cleaned['unit_price'] - df_cleaned['comp_3']
```

```python
# Calculate total revenue (total_price = unit_price * qty)
df_cleaned['total_revenue'] = df_cleaned['unit_price'] * df_cleaned['qty']
```

```python
# Bar graph: Unit Price vs Competitor Prices for the first few products
df_bar = df_cleaned[['product_id', 'unit_price', 'comp_1', 'comp_2', 'comp_3']].head(10)
df_bar.set_index('product_id', inplace=True)
```

```python
#Price Difference: Between your store's price and competitors' prices (comp_1, comp_2, comp_3).
```

```python
# Plot the bar graph
df_bar.plot(kind='bar', figsize=(10,6))
plt.title('Unit Price vs Competitor Prices')
plt.ylabel('Price')
plt.xlabel('Product ID')
plt.xticks(rotation=45)
plt.legend(title='Price Type')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot: Sales Volume vs Unit Price
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['unit_price'], df_cleaned['qty'], alpha=0.6, color='b')
plt.title('Sales Volume vs Unit Price')
plt.xlabel('Unit Price')
plt.ylabel('Quantity Sold')
plt.grid(True)
plt.show()
```

```python
#2. Top 10 Products by Total Revenue (Bar Graph)
```

```python
# Bar graph: Total Revenue by Product
df_revenue = df_cleaned[['product_id', 'total_revenue']].groupby('product_id').sum().sort_values(by='total_revenue', ascending=False).head(10)

# Plot the bar graph
df_revenue.plot(kind='bar', figsize=(10,6))
plt.title('Top 10 Products by Total Revenue')
plt.ylabel('Total Revenue')
plt.xlabel('Product ID')
plt.xticks(rotation=45)
plt.legend(title='Revenue')
plt.show()
```

```python
import matplotlib.pyplot as plt

df_cleaned['total_price'].hist()
plt.title('Distribution of Total Price')
plt.xlabel('Total Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Convert 'month_year' column to datetime
df_cleaned['month_year'] = pd.to_datetime(df_cleaned['month_year'], format='%d-%m-%Y')
```

```python
df_cleaned['month_year'] = pd.to_datetime(df_cleaned['month_year'], format='%Y-%m')
df_cleaned.set_index('month_year', inplace=True)

df_cleaned['total_price'].resample('M').sum().plot()
plt.title('Total Price Over Time')
plt.xlabel('Month')
plt.ylabel('Total Price')
plt.show()
```

```python
#1. Estimate Financial Impact
df_cleaned['price_change'] = df_cleaned['unit_price'] - df_cleaned['lag_price']
df_cleaned['financial_impact'] = df_cleaned['price_change'] * df_cleaned['qty']
print(df_cleaned[['product_id', 'price_change', 'financial_impact']].head())
```

```python
#1. Financial Impact by Product
#To visualize the financial impact of different products:
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure data is sorted for better visualization
price_data_sorted = df_cleaned.sort_values(by='financial_impact', ascending=False)

# Set the plot size
plt.figure(figsize=(12, 8))

# Create a barplot using seaborn
sns.barplot(x='product_id', y='financial_impact', data=price_data_sorted, palette='viridis')

# Add titles and labels
plt.title('Financial Impact by Product', fontsize=16)
plt.xlabel('Product ID', fontsize=14)
plt.ylabel('Financial Impact', fontsize=14)

# Rotate the x-ticks for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()

```

```python
# Create 'month_year' by combining 'month' and 'year'
# Convert 'month_year' column to datetime
df_cleaned['month_year'] = pd.to_datetime(df_cleaned['month_year'], format='%d-%m-%Y')

# Aggregate sales volume by month
monthly_volume = df_cleaned.groupby(df_cleaned['month_year'].dt.to_period('M'))['volume'].sum().reset_index()
monthly_volume['month_year'] = monthly_volume['month_year'].astype(str)  # Convert period to string for plotting

# Plot the bar graph
plt.figure(figsize=(12, 8))
sns.barplot(x='month_year', y='volume', data=monthly_volume, palette='coolwarm')
plt.title('Sales Volume by Month')
plt.xlabel('Month-Year')
plt.ylabel('Sales Volume')
plt.xticks(rotation=45)
plt.show()

```

```python
# Prepare Store Retail Adjustments
# Assuming you have a column for adjustment percentage
price_data['adjusted_price'] = price_data['unit_price'] * (1 + price_data['price_change_percentage']/100)
price_data['projected_impact'] = (price_data['adjusted_price'] - price_data['unit_price']) * price_data['qty']
print(price_data[['product_id', 'unit_price', 'adjusted_price', 'projected_impact']].head())
```

```python
#Analyze Strategic Pricing Initiatives
# Group by product_id to assess impact
item_level_impact = price_data.groupby('product_id').agg({
    'total_price': 'sum',
    'financial_impact': 'sum',
    'qty': 'sum'
}).reset_index()

# Calculate margins
item_level_impact['margin'] = item_level_impact['total_price'] - item_level_impact['financial_impact']
print(item_level_impact)
```

```python
 #Price Indexing and Margin Assessment
# Calculate price index and margin
price_data['price_index'] = price_data['unit_price'] / price_data['unit_price'].mean()
price_data['margin'] = price_data['total_price'] - (price_data['unit_price'] * price_data['qty'])
print(price_data[['product_id', 'price_index', 'margin']].head())
```
